# Random Enrollment DataFrames Generator
Generates **3 DataFrames** of 500 rows each with randomized `student_id` and `course_id` (range 1–566).

```python
import pandas as pd
import numpy as np

students = pd.read_csv('/Users/dazedinthecity/Documents/GitHub/CEU-DMDB-GroupH/data/private/real_data/student.csv')

NUM_ROWS = 500
rng = np.random.default_rng(seed=42)  # remove seed for non-reproducible results

def make_enrollment_df(rng):
    return pd.DataFrame({
        'index':      range(1, NUM_ROWS + 1),
        'student_id': rng.choice(students['student_id'].values, size=NUM_ROWS, replace=True),
        'course_id':  rng.integers(1, 567, size=NUM_ROWS),  # random int in range 1-566
    })

df1 = make_enrollment_df(rng)
df2 = make_enrollment_df(rng)
df3 = make_enrollment_df(rng)

print('df1:', df1.shape)
print('df2:', df2.shape)
print('df3:', df3.shape)
```

```python
# Save all 3 DataFrames to CSV
df1.to_csv('enrollments_1.csv', index=False)
df2.to_csv('enrollments_2.csv', index=False)
df3.to_csv('enrollments_3.csv', index=False)

print('Saved: enrollments_1.csv, enrollments_2.csv, enrollments_3.csv')
```
