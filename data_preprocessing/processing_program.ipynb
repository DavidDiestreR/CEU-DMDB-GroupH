```python
import random

first_names = ['james', 'emma', 'luca', 'sofia', 'noah', 'olivia', 'ethan', 'mia', 'alex', 'sara']
last_names = ['smith', 'brown', 'jones', 'garcia', 'miller', 'davis', 'wilson', 'taylor', 'moore', 'anderson']

random.seed(42)
all_combos = [f"{f}_{l}@ceu.edu" for f in first_names for l in last_names]  # 100 unique combos
random.shuffle(all_combos)

df_all['email'] = all_combos[:len(df_all)]
```

```python
import random
import string

random.seed(42)

# --- Program codes ---
def generate_program_code():
    letters = ''.join(random.choices(string.ascii_uppercase, k=3))
    numbers = ''.join(random.choices(string.digits, k=3))
    return f"{letters}{numbers}"

codes = random.sample([generate_program_code() for _ in range(500)], 79)
df_all['program_code'] = codes

# --- Descriptions & Learning Outcomes ---
disciplines = ['quantitative', 'interdisciplinary', 'applied', 'theoretical', 'comparative', 'empirical', 'critical']
focuses = ['policy analysis', 'research methods', 'data-driven inquiry', 'global challenges', 'social change', 'governance', 'innovation']
skills = ['analytical thinking', 'academic writing', 'statistical modeling', 'fieldwork', 'critical reasoning', 'ethical judgment', 'cross-cultural communication']
tools = ['R', 'Python', 'GIS software', 'SPSS', 'qualitative methods', 'archival research', 'survey design']
outcomes = ['conduct independent research', 'apply theoretical frameworks', 'engage with real-world problems',
            'critically evaluate evidence', 'communicate findings to diverse audiences', 'collaborate across disciplines',
            'design and implement research projects']

descriptions = []
learning_outcomes = []

for program in df_all['program_name']:
    desc = (
        f"This program offers a {random.choice(disciplines)} approach to {random.choice(focuses)}, "
        f"equipping students with skills in {random.choice(skills)} and {random.choice(skills)}. "
        f"Students engage with {random.choice(focuses)} through {random.choice(tools)} and {random.choice(tools)}, "
        f"preparing them for careers in academia, policy, and the private sector."
    )

    lo = (
        f"By the end of this program, students will be able to: "
        f"(1) {random.choice(outcomes)}, "
        f"(2) {random.choice(outcomes)}, "
        f"(3) {random.choice(outcomes)}, "
        f"(4) apply {random.choice(tools)} in the context of {random.choice(focuses)}, "
        f"and (5) demonstrate proficiency in {random.choice(skills)}."
    )

    descriptions.append(desc)
    learning_outcomes.append(lo)

df_all['program_description'] = descriptions
df_all['learning_outcomes'] = learning_outcomes
```
