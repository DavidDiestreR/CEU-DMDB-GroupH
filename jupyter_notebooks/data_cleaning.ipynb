```python
from pathlib import Path

repo_root = Path.cwd().resolve()
if not (repo_root / 'data').exists():
    repo_root = repo_root.parent

file_path_TE = repo_root / 'data' / 'private' / 'TE_Course_Data_25-26.csv'

df_TE = pd.read_csv(file_path_TE)

df_TE.shape
```

## Dealing with NAs

```python
# Missing-value summaries for df_TE
# per-column missing counts and percent
missing_col_TE = df_TE.isna().sum()
missing_col_pct_TE = (df_TE.isna().mean() * 100).round(2)
missing_col_TE_df = pd.DataFrame({'missing_count': missing_col_TE, 'missing_pct': missing_col_pct_TE})

display(missing_col_TE_df.sort_values('missing_count', ascending=False))

# per-row missing counts and percent (show top rows with most missing values)
missing_row_TE = df_TE.isna().sum(axis=1)
missing_row_pct_TE = (df_TE.isna().mean(axis=1) * 100).round(2)
missing_row_TE_df = pd.DataFrame({'missing_count': missing_row_TE, 'missing_pct': missing_row_pct_TE})

display(missing_row_TE_df.sort_values('missing_count', ascending=False).head(20))

# quick totals
print('Rows with any missing values (df_TE):', df_TE.isna().any(axis=1).sum())
print('Total cells (df_TE):', df_TE.size)
print('Total missing cells (df_TE):', df_TE.isna().sum().sum())
```

```python

# per-column missing counts and percent
missing_col = df.isna().sum()
missing_col_pct = (df.isna().mean() * 100).round(2)
missing_col_df = pd.DataFrame({'missing_count': missing_col, 'missing_pct': missing_col_pct})

display(missing_col_df.sort_values('missing_count', ascending=False))

# per-row missing counts and percent (show top rows with most missing values)
missing_row = df.isna().sum(axis=1)
missing_row_pct = (df.isna().mean(axis=1) * 100).round(2)
missing_row_df = pd.DataFrame({'missing_count': missing_row, 'missing_pct': missing_row_pct})

display(missing_row_df.sort_values('missing_count', ascending=False).head(20))

# quick totals
print('Rows with any missing values:', df.isna().any(axis=1).sum())
print('Total cells:', df.size)
print('Total missing cells:', df.isna().sum().sum())
```

```python
# Drop columns with >=50% missing values for both df and df_TE
thresh = 50.0

# For df
missing_col_pct = (df.isna().mean() * 100).round(2)
cols_to_drop_df = missing_col_pct[missing_col_pct >= thresh].index.tolist()
print('Columns to drop from df (>=50% missing):', cols_to_drop_df)
if cols_to_drop_df:
    df.drop(columns=cols_to_drop_df, inplace=True)
print('df shape after drop:', df.shape)

# For df_TE
missing_col_pct_TE = (df_TE.isna().mean() * 100).round(2)
cols_to_drop_df_TE = missing_col_pct_TE[missing_col_pct_TE >= thresh].index.tolist()
print('Columns to drop from df_TE (>=50% missing):', cols_to_drop_df_TE)
if cols_to_drop_df_TE:
    df_TE.drop(columns=cols_to_drop_df_TE, inplace=True)
print('df_TE shape after drop:', df_TE.shape)
```

## Combining the two dataframes

```python
# Combine `df` and `df_TE` by concatenation (aligns columns automatically)
# show common and unique columns first
common_cols = df.columns.intersection(df_TE.columns).tolist()
only_df = df.columns.difference(df_TE.columns).tolist()
only_df_TE = df_TE.columns.difference(df.columns).tolist()

print('Common columns:', common_cols)
print('Columns only in df:', only_df)
print('Columns only in df_TE:', only_df_TE)

# Concatenate (outer alignment of columns)
combined_df = pd.concat([df, df_TE], ignore_index=True, sort=False)
print('combined_df shape:', combined_df.shape)

# quick peek
combined_df.head()
```
